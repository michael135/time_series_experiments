```python
df = pd.read_csv('train.csv', parse_dates=['date'], index_col='date')
```

```python
df[df['store']==1]['sales'].loc['2013-01':'2013-03'].plot(figsize=(10,5));
```

```python
df['item'].value_counts()
```

```python
df_1_1 = df[(df['store']==1) & (df['item']==1)][['sales']]
df_10_10 = df[(df['store']==5) & (df['item']==10)][['sales']]
df_30_30 = df[(df['store']==10) & (df['item']==30)][['sales']]
```

```python
del df_1_1['sma2']
```

```python
df_1_1['2013'].plot(figsize=(13, 4));
df_10_10.plot(figsize=(13, 4));
df_30_30.plot(figsize=(13, 4));
```

```python
df_1_1['sma2'] = df_1_1['sales'].rolling(2).mean()
df_1_1['sma7'] = df_1_1['sales'].rolling(7).mean()
df_1_1['sma14'] = df_1_1['sales'].rolling(14).mean()
df_1_1['sma30'] = df_1_1['sales'].rolling(30).mean()
```
